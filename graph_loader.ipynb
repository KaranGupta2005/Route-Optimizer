```python
import osmnx as ox
```

```python
def load_graph(city="Delhi, India"):
    print("Downloading graph network...")
    graph = ox.graph_from_place(city, network_type="drive")
    return graph
```

```python
if __name__ == "__main__":
    graph = load_graph()
    print("Nodes:", len(graph.nodes))
    print("Edges:", len(graph.edges))
```
