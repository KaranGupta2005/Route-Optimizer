```python
import json,random
```

```python
def generate(n=5000):
    data=[]

    for i in range(n):
        rainfall=random.uniform(0,120)
        flood=random.randint(0,10)
        speed=random.uniform(5,60)
        elevation=random.uniform(200,260)

        risk=( 0.4*(rainfall/120) + 0.3*(flood/10) + 0.2*(1-speed/60) + 0.1*(1 if elevation<220 else 0))

        data.append({ "length":random.uniform(50,800),"road_type":random.choice([0,1,2]),"elevation":elevation,"flood_reports":flood,"traffic_speed":speed,"is_night": random.randint(0,1),
            "risk_score": min(risk,1) })

    with open("road_data.json","w") as f:
        json.dump(data,f,indent=2)         
```

```python
generate()
```
